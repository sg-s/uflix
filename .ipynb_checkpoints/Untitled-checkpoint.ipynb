```python
import os
import io
import re
import time
import subprocess
from guessit import guessit

from imdbpie import Imdb
imdb = Imdb(anonymize=True) # to proxy requests

path_name = '/Volumes/SKADI/movies'
dirs = [d for d in os.listdir(path_name) if os.path.isdir(os.path.join(path_name, d))]

# define a set of strings to remove from the folder_name before searching
bad_strings = ['aac','BDRip','subs','ac3','mp4','D1','Disc_3','Disc_2','Disc_1','HDrip','[',']','720p','1080p','1080i','DVDrip','x264','Full Movie','HD','H264','.','_','BluRay','dvd','XviD','divx','BRRip']

# define sleep time -- if too short, Google will block your IP!
sleep_time = 10 # seconds

# create a text file inside each folder with the original name of the folder. 
# this way, in case we screw up the rename, we can always undo

print('Backing up folder names...')
for i in range(0,len(dirs)):
    if dirs[0][0] == '.':
        continue
    
    
    # check for a "original_name.uflix" in the folder
    txt_file = os.path.join(path_name,dirs[i],'original_name.uflix')
    if not os.path.exists(txt_file):
        with io.FileIO(txt_file, "w") as file:
            file.write(dirs[i].encode())
            file.close()
            
print("Guessing correct names and clearning up names...")
bad_title = []
bad_year = []
could_not_rename = []
for i in range(0,len(dirs)):
    if dirs[i][0] == '.':
        continue
    
#     print(dirs[i], end='') 
#     print('   ---->   ', end='') 

    this_string = dirs[i]
    
    g = guessit(this_string)
    
    new_name = this_string
    if ("year" in g.keys() and "title" in g.keys()):
        new_name = g['title'] + " (" + str(g['year']) + ")"
    elif ("title" in g.keys()):
        bad_year.append(this_string)
        new_name = g['title']
    else:
        bad_title.append(this_string)
        
    new_name = new_name.title()
#     print(new_name)    
    
    # now the actual work -- rename the folder
    old_name = os.path.join(path_name,dirs[i])
    new_name = os.path.join(path_name,new_name)
    
    
    try:
        os.rename(old_name,new_name)
    except: 
        print("Error in renaming folder. Probably because you had a folder with the same name already")
        could_not_rename.append(new_name)
    

# OK, now, search IMDB for the ones that we couldn't parse propoerly
not_on_imdb = []
print("Searching IMDB for year info...")
for this_string in bad_year:    
    r = imdb.search_for_title(this_string)

    new_name = this_string
    if len(r) > 0:
        if r[0]['year'] == None:
            not_on_imdb.append(this_string)
        else:
            new_name = r[0]['title'] + " (" + r[0]['year'] + ")"
    else:
        not_on_imdb.append(this_string)
        
#     print(this_string, end='') 
#     print('   ---->   ', end='') 
#     print(new_name)
        
    # now the actual work -- rename the folder
    old_name = os.path.join(path_name,this_string)
    new_name = os.path.join(path_name,new_name)
    
    try:
        os.rename(old_name,new_name)
    except: 
        print("Error in renaming folder. Probably because you had a folder with the same name already")
        could_not_rename.append(new_name)
    
    
    

#     for s in bad_strings:
#         insensitive_hippo = re.compile(re.escape(s), re.IGNORECASE)
#         this_string = insensitive_hippo.sub(' ', this_string)
       

    
#     process = subprocess.Popen(['googler','-C','-n','1','-w','imdb.com','--np',this_string], stdout=subprocess.PIPE)
#     out, err = process.communicate()
#     out = out.decode(encoding='UTF-8')
#     imdb_name = out[3:out.find('- IMDb')]
    
#     if len(imdb_name) > 50:
#         print("Name too long, skipping...")
#         bad_files.append(i)
#         continue

    
#     if imdb_name.find('(') < 0:
#         print("Wrong format, skipping...")
#         bad_files.append(i)
#         continue

    
#     if imdb_name.find(')') < 0:
#         print("Wrong format, skipping...")
#         bad_files.append(i)
#         continue
    
#     print(imdb_name)
#     time.sleep(sleep_time)
                                
print('All done')
```
